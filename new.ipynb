```python
import pandas as pd
import numpy as np

file = pd.read_csv(r"C:\Users\USER\Desktop\DATASCIENCE\project\Economy\Nike_Sales_Uncleaned.csv")
file.dropna(subset=["Size"])

file.loc[~file["Size"].str.isalpha(),"Size"] = np.nan
# print(file["Size"].isnull().sum())
file.dropna()
file = file[file["Size"].notna()]
print(file["Size"])
file.to_csv(r"C:\Users\USER\Desktop\DATASCIENCE\project\Economy\Nike_Sales_Size_Cleaned.csv", index=False)

```
